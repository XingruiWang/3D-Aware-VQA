```python
from collections import defaultdict
import math

# Function to load OFF file
def load_off_file(filename):
    with open(filename, 'r') as f:
        lines = f.readlines()

    lines = [line.strip().split() for line in lines]

    if lines[0][0] != 'OFF':
        print("Not a valid OFF header")
        return None, None

    n_vertices, n_faces, _ = map(int, lines[1])

    vertices = [list(map(float, line)) for line in lines[2:2 + n_vertices]]
    faces = [list(map(int, line[1:])) for line in lines[2 + n_vertices:2 + n_vertices + n_faces]]

    return vertices, faces

# Function to build adjacency list from faces
def build_adjacency_list(faces):
    adjacency_list = defaultdict(set)
    for face in faces:
        for i in range(len(face)):
            for j in range(i + 1, len(face)):
                adjacency_list[face[i]].add(face[j])
                adjacency_list[face[j]].add(face[i])
    return adjacency_list

# Calculate Euclidean distance between two vertices
def distance(v1, v2):
    return math.sqrt(sum((x - y) ** 2 for x, y in zip(v1, v2)))

# Function to sort vertices using DFS and distance for disconnected components
def dfs_sort(vertices, adjacency_list):
    visited = set()
    sorted_indices = []

    def dfs(vertex):
        nonlocal sorted_indices
        if vertex in visited:
            return
        visited.add(vertex)
        sorted_indices.append(vertex)
        for neighbor in adjacency_list[vertex]:
            dfs(neighbor)

    # Start DFS from the first vertex
    dfs(0)

    # Continue DFS from closest unvisited vertex for disconnected components
    while len(visited) < len(vertices):
        min_distance = float('inf')
        closest_vertex = None
        last_vertex = sorted_indices[-1]

        for i in range(len(vertices)):
            if i not in visited:
                dist = distance(vertices[last_vertex], vertices[i])
                if dist < min_distance:
                    min_distance = dist
                    closest_vertex = i

        dfs(closest_vertex)

    return sorted_indices

# Main function to sort vertex indices based on connectivity
def sort_vertex_indices_by_connectivity(file_path):
    vertices, faces = load_off_file(file_path)
    if vertices is None or faces is None:
        return None

    adjacency_list = build_adjacency_list(faces)
    sorted_indices = dfs_sort(vertices, adjacency_list)
    return sorted_indices
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os


file_name = 31880
category = 'motorbike'

# Create output directory
output_dir = f'frames_{file_name}'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

data = np.load(f'/home/xingrui/publish/PO3D-VQA/6D_pose_estimator/feature_{file_name}.npy')
sorted_indices = sort_vertex_indices_by_connectivity(f'/home/xingrui/publish/PO3D-VQA/6D_pose_estimator/CAD_cate/{category}/01.off')


# Read the image you want to overlay
overlay_image = plt.imread(f'/home/xingrui/vqa/super-clevr-gen/output/ver_mask_new/images/superCLEVR_new_{file_name:06d}.png')

# Number of slices
num_slices = data.shape[0]

# Initialize figure
fig, ax = plt.subplots()

# Loop through slices
for i in range(num_slices):
    # Display overlay image
    ax.imshow(overlay_image, extent=[0, 80, 0, 60])
    
    # Display mask (from 3D array)
#     ax.imshow(data[:, :, i], cmap='gray', alpha=0.5, extent=[0, 80, 0, 60])
    ax.imshow(data[sorted_indices[i], :, :], cmap='viridis', alpha=0.5,  extent=[0, 80, 0, 60], vmin=0, vmax=1)

    ax.set_title(f'Keypoint {i}')
    
    # Save frame as image
    plt.savefig(f"{output_dir}/frame_{i:03d}.png")
    
    ax.clear()
```

```python
file_name = 31880
data = np.load(f'/home/xingrui/publish/PO3D-VQA/6D_pose_estimator/feature_{file_name}.npy')
num_slices = data.shape[0]
# Initialize video writer
output_dir = f'frames_{file_name}'
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(f'slices_video_{file_name}.mp4', fourcc, 8.0, (640, 480))

# Compile frames into video
for i in range(num_slices):
    img_path = f"{output_dir}/frame_{i:03d}.png"
    frame = cv2.imread(img_path)
    
    # Resize frame if needed
    frame_resized = cv2.resize(frame, (640, 480))
    
    out.write(frame_resized)

# Release video writer
out.release()
```
